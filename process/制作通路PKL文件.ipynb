# 制作1654个通路PKL文件

```python
import pickle
import pandas as pd
```

```python
# 读取1654个通路文件
df_pathways = pd.read_csv('../KEGG_data/OmicsExpressionProteinCodingGenesTPMLogp1-23Q2_Only_MEDICUS_GSVA.csv',header=0).columns.values.tolist()
# 去掉第一个元素cell_line
df_pathways = df_pathways[1:]
```

```python
# 将df_pathways写入pkl文件
with open('../KEGG_data/MUDICUS_Omic_619_pathways.pkl', 'wb') as f:
    pickle.dump(df_pathways, f)
    
```

```python

# 读取PKL文件
with open('../KEGG_data/MUDICUS_Omic_619_pathways.pkl', 'rb') as f:
    data = pickle.load(f)
    print(len(data))
    print(data)
```
