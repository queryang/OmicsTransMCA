# 基因表达秩和检验P值转换

```python
import pandas as pd
from math import log  
```

```python
df_gene_p_value = pd.read_csv('../KEGG_data/GeneExp_Wilcoxon_test_Analysis_Log10_P_value_C2_KEGG186.csv', index_col=0)
df_gene_p_value.head()
```

```python
# 将df_gene_p_value所有值转换为-log10(P)
df_gene_p_value = -df_gene_p_value.map(lambda x: log(x,10))
df_gene_p_value.head()
```

```python
df_gene_p_value.to_csv('../KEGG_data/GeneExp_Wilcoxon_test_Analysis_Log10_P_value_C2_KEGG186.csv')
```
