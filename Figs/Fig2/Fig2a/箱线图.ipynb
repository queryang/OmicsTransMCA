# Fig2A: 各肺癌亚型细胞系 箱线图

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
import pandas as pd
%matplotlib inline
plt.style.use('default')
```

```python
df_data = pd.read_csv("../../data_NON_NSCLC/df_LUNG_Filtered_ZScore.csv")
# 按照Cell Lines排序
df_data = df_data.sort_values(by="OncotreeCode")
df_data.head()
```

```python
palette = sns.color_palette("Set2", 5)
box_plot = sns.boxplot(data=df_data,y="Z-Score",x="cell_line",hue="OncotreeCode",palette=palette, linewidth=1.6)
# 设置x轴标签倾斜45度
plt.xticks(rotation=45)
# box_plot.set_xticklabels(['LUSC', 'LUAD', 'LCLC', 'SCLC', 'NSCLC'], rotation=45)
# 将图例放在图的右边
box_plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Get the current axes, creating one if necessary.
ax = plt.gca()
# Set the spines (the box) visibility
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['bottom'].set_linewidth(2)  # X轴线宽
ax.spines['left'].set_linewidth(1)  # Y轴线宽
ax.spines['bottom'].set_linewidth(1)

# 让箱线图稍微右移一点，不要跟y轴重合
box_plot.set_position([0.1, 0.1, 0.9, 0.8])

# plt.tight_layout()

plt.show()
```

```python
# 画出df_data的小提琴图
plt.figure(figsize=(12, 10))
# 设置全局字体
plt.rcParams['font.sans-serif'] = ['Bahnschrift']
plt.rcParams['axes.unicode_minus'] = False
# 设置颜色方案
# palette = sns.color_palette("muted", 5)
# 'LUAD','LCLC','NSCLC','LUSC','SCLC'
palette = {'LUAD':'#377EB8',
            'LCLC':'#4DAF4A',
            # 'NSCLC':'#ff9f9b',
            'LUSC':'#E41A1C',
            'SCLC':'#FF7F00'}
# "PRE-CX"="#984EA3","POST-CX"="#FF7F00",CRPC="#E41A1C",ENZS="#4DAF4A",ENZR="#377EB8"
# 创建小提琴图
violin_plot = sns.violinplot(data=df_data, y="Z-Score", x="cell_line", hue="OncotreeCode", palette=palette, linewidth=1.8,bw_adjust=.5, cut=1,)
# Illustrating strip plot on top of box plot
# sns.stripplot(data=df_data, y="Z-Score", x="cell_line", hue="OncotreeCode", palette=palette, dodge=False, jitter=True, alpha=0.7)

# 设置x轴标签倾斜45度
plt.xticks(rotation=55, fontsize=12)
plt.yticks(fontsize=12)
# 设置标题和轴标签
# plt.title('Boxplot of Predicted Z-Score for Different Cell Lines', fontsize=20)
plt.xlabel('Cell Lines', fontsize=15)
plt.ylabel('Predicted LN IC50 (Z-Score)', fontsize=16)

# 将图例放在图的右边，并设置图例的大小

violin_plot.legend(loc='lower left', title='Subtype',edgecolor='black',fontsize=10, title_fontsize=10, shadow=False, fancybox=True, frameon=True, framealpha=1, bbox_to_anchor=(0.001, 0.001))

# Get the current axes, creating one if necessary.
ax = plt.gca()
# Set the spines (the box) visibility
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['bottom'].set_linewidth(2)  # X轴线宽
# ax.spines['left'].set_linewidth(1)  # Y轴线宽
# ax.spines['bottom'].set_linewidth(1)

# 让箱线图稍微右移一点，不要跟y轴重合
# box_plot.set_position([0.1, 0.1, 0.9, 0.8])
# 添加网格线
# plt.grid(True)
# plt.tight_layout()
plt.savefig('violin_plot.png',dpi = 300, bbox_inches = 'tight')
# 显示图形
plt.show()
```

```python
# 画出箱线图，展示不同组学数据的MSE分布
plt.figure(figsize=(10, 10))
# scatter_props = dict(color='k', marker='o', alpha=0.5)
# 绘制箱线图,并自定义样式
box_props = dict(facecolor='#5DADE2', edgecolor='#2C3E50', linewidth=2,) # 设置箱体边框颜色和线宽
whisker_props = dict(color='k', linestyle='-', linewidth=2,alpha=0.8) # 设置whisker线条颜色、样式和线宽
cap_props = dict(color='k', linewidth=2,alpha=0.8)  # 设置上方横线的颜色和线宽
median_props = dict(color='k', linewidth=2) # 设置中位数线条颜色和线宽
ax = sns.boxplot(x='cell_line', y='prediction', data=df_data,hue="OncotreeCode",linewidth=2, 
                 boxprops=box_props, whiskerprops=whisker_props, medianprops=median_props,
                 showfliers=False, width=0.5,capprops=cap_props,palette=palette)
# Add in points to show each observation
sns.stripplot(x='cell_line', y='prediction', data=df_data, jitter=True, color='k', alpha=0.7, size=6)

# plt.title('MSE of Different Genomic Data', fontsize=25)
plt.ylabel('Predicted LN IC50', fontsize=18)
plt.xlabel('Genomic Data', fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=18)
# 去除上和右边框
# plt.gca().spines['top'].set_visible(False)
# plt.gca().spines['right'].set_visible(False)
# plt.gca().spines['bottom'].set_visible(False)
# 去除下方的刻度
# plt.gca().xaxis.set_ticks_position('none')
# 加粗横纵坐标
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
# 刻度加粗
plt.gca().tick_params(width=2, length=6)

# plt.savefig('MSE of Different Genomic Data.png', dpi=300)
plt.show()
```
