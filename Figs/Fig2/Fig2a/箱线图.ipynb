# Fig2A: 各肺癌亚型细胞系 箱线图

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors
import pandas as pd
%matplotlib inline
plt.style.use('default')
```

```python
df_data = pd.read_csv("../../data/LUNG_Filtered_ZScore.csv")
# 按照Cell Lines排序
df_data = df_data.sort_values(by="OncotreeCode")
df_data.head()
```

```python
palette = sns.color_palette("Set2", 5)
box_plot = sns.boxplot(data=df_data,y="Z-Score",x="cell_line",hue="OncotreeCode",palette=palette, linewidth=1.6)
# 设置x轴标签倾斜45度
plt.xticks(rotation=45)
# box_plot.set_xticklabels(['LUSC', 'LUAD', 'LCLC', 'SCLC', 'NSCLC'], rotation=45)
# 将图例放在图的右边
box_plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Get the current axes, creating one if necessary.
ax = plt.gca()
# Set the spines (the box) visibility
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['bottom'].set_linewidth(2)  # X轴线宽
ax.spines['left'].set_linewidth(1)  # Y轴线宽
ax.spines['bottom'].set_linewidth(1)

# 让箱线图稍微右移一点，不要跟y轴重合
box_plot.set_position([0.1, 0.1, 0.9, 0.8])

# plt.tight_layout()

plt.show()
```

```python
# 设置全局字体
plt.rcParams['font.sans-serif'] = ['Bahnschrift']
plt.rcParams['axes.unicode_minus'] = False
# 设置颜色方案
# palette = sns.color_palette("muted", 5)
# 'LUAD','LCLC','NSCLC','LUSC','SCLC'
palette = {'LUAD':'#b9f2f0',
           'LCLC':'#d0bbff',
           'NSCLC':'#ff9f9b',
           'LUSC':'#a6d854',
           'SCLC':'#66c2a5'}
# 设置全局字体
plt.rcParams['font.sans-serif'] = ['Bahnschrift']
plt.rcParams['axes.unicode_minus'] = False
# 创建箱线图
box_plot = sns.boxplot(data=df_data, y="Z-Score", x="cell_line", hue="OncotreeCode", palette=palette, linewidth=1.4)

# 设置x轴标签倾斜45度
plt.xticks(rotation=60)

# 设置标题和轴标签
plt.title('Boxplot of Predicted Z-Score for Different Cell Lines', fontsize=15)
plt.xlabel('Cell Lines', fontsize=12)
plt.ylabel('Predicted LN IC50 (Z-Score)', fontsize=12)

# 将图例放在图的右边，并设置图例的大小

box_plot.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Subtype')

# Get the current axes, creating one if necessary.
ax = plt.gca()
# Set the spines (the box) visibility
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['bottom'].set_linewidth(2)  # X轴线宽
ax.spines['left'].set_linewidth(1)  # Y轴线宽
ax.spines['bottom'].set_linewidth(1)

# 让箱线图稍微右移一点，不要跟y轴重合
box_plot.set_position([0.1, 0.1, 0.9, 0.8])
# 添加网格线
# plt.grid(True)
plt.tight_layout()
plt.savefig('box_plot.png',dpi = 200)
# 显示图形
plt.show()
# plt.close()
```
